```python
import pandas as pd
```

```python
df = pd.read_csv('../hdb_2605.csv')
```

```python
df['Project'] = df['Project'].fillna(method='ffill')
```

```python
sel = df['Flat_type'].str.contains('4-room|5-room', na=False)
df = df.loc[sel].copy()
```

```python
sel = (df['Flat_type'] != '4-room')

df.loc[sel,'Flat_type'] = '5-room'
```

```python
df.sort_values('first_timer', ascending=False).query('Flat_type == "4-room"')
```

```python
df.sort_values('first_timer').query('Flat_type != "4-room"')
```

```python
# Toa Payoh, Punggol, Tampines, Sengkang, Tengah, Yishun
```

```python
shortlisted = (
    df
    .groupby('Project')
    ['Num_units']
    .sum()
    .sort_values()
    .tail(6)
    .index
)
```

```python
sel = df['Project'].isin(shortlisted)

(
    df
    .loc[sel]
    .set_index(['Project','Flat_type'])
    ['Num_units']
    .unstack(-1)
    .
)
```
